```python
# EXTRACTING THE COLUMNS FROM THE DATAFRAME
columns = df.columns.tolist()
columns, len(columns)
```

```python
def drop_columns(df):
    '''
    This function drops columns with perecntage null values greater than 10
    '''
    columns = df.columns.tolist()
    for col in columns:
        total_val_num = len(df[col])
        null_val_num=df[col].isnull().sum()
        percentage_null = (null_val_num/ total_val_num) * 100
        if percentage_null>=10:
            df.drop(col, axis=1, inplace=True)
    return df
```

```python
## filling nan values with median values for numerical columns:
def fill_median_or_mean_value(df, num_columns):
    for col in num_columns:
        df1 = df[col]
        df[col].fillna(np.mean(df1), inplace = True)
fill_median_or_mean_value(df, num_columns)
```

```python
def removing_categorical_column_as_per_variance(df):
    dropped_columns = []
    for col in df.columns:
        if len(df[col].unique())>1:
            unique_obj_per = np.array(df[col].value_counts(normalize = True)* 100)
            var = np.var(unique_obj_per)
            if var<1:
                dropped_columns.append(col)
        else:
            dropped_columns.append(col)
    df.drop(dropped_columns, axis='columns', inplace=True)
removing_categorical_column_as_per_variance(cat_df)
cat_df.sample(5)
```

```python
# Onehot encoding of the categorical dataframe
cat_df_columns = cat_df.columns.tolist()
cat_df = pd.get_dummies(cat_df, drop_first=True)
cat_df.sample(5)
```

```python
X=pd.concat([num_df, cat_df],axis=1)
X.sample(5)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

```python
#Logistic regression

from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import cross_val_predict 
from sklearn.metrics import confusion_matrix, f1_score 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
penalty = ['l1', 'l2']
C = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
hyper_params = []
f1_Scores = []
```

```python
lgr_best_score = 0
lgr_best_solver = str
lgr_best_c = 0
lgr_best_penalty = str
kFolds = 5
```

```python
for s in solver:
    for p in penalty:
        for c in C:
            
            if s == 'newton-cg' or 'lbfgs':
                
                if p != 'l1':
                    
                    lgr = LogisticRegression(penalty = p , C = c,  solver = s, max_iter = 100)
                    y_pred = cross_val_predict(lgr, x_train, y_train, cv = kFolds)
                    F1_score = f1_score(y_train, y_pred)
                    c_matrix = confusion_matrix(y_train, y_pred)
                    hyper_params.append([s,p,c]) 
                    f1_Scores.append(F1_score)
                    
                    if F1_score >= lgr_best_score:
                        lgr_best_score = F1_score
                        lgr_best_solver = s
                        lgr_best_c =c
                        lgr_best_penalty = p
                        
            else:
                

                lgr = LogisticRegression(penalty = p, C =c,  solver = s, max_iter = 100)
                y_pred = cross_val_predict(lgr, x_train, y_train, cv = kFolds)
                F1_score = f1_score(y_train, y_pred)
                c_matrix = confusion_matrix(y_train, y_pred)
                hyper_params.append([s,p,c]) 
                f1_Scores.append(F1_score)
                    
                if F1_score >= lgr_best_score:
                    lgr_best_score = F1_score
                    lgr_best_solver = s
                    lgr_best_c = c
                    lgr_best_penalty = p
```
