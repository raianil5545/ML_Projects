```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("AmesHousing.csv")
df.sample(5)
```

```python
# Exploratory Data Analysis 
df.info()
```

```python
# dropping the columns with more than 20 percent nan values
def drop_columns_with_greater_percentage_nan(df):
    total_records = df.shape[0]
    dropped_cols = []
    for col in df.columns.to_list():
        null_records = df[col].isna().sum()
        null_records_pct = (null_records/total_records) * 100
        if null_records_pct > 20:
            dropped_cols.append(col)
    return dropped_cols

cols_drop = drop_columns_with_greater_percentage_nan(df)
print("Columns dropped: ", cols_drop)
df.drop(columns = cols_drop, axis=1, inplace=True)
        
```

```python
for col in df.columns.to_list():
    print(f"{col} has {len(df[col].unique())} unique values")
```

```python
# No missing target variables
df.SalePrice.isna().sum()
```

```python
# we are dropping the columns ['Order', 'PID', 'Lot Area']
df.drop(columns=['Order', 'PID', 'Lot Area'], axis=1, inplace=True)
df.sample(5)
```

```python
# Missing value imputation for the numerical features and dropping the records for missing categorical values
def fill_with_mean_values(X):
    for col in df.columns:
        if df[col].dtypes in ['float64', 'int64']:
            df[col].fillna(np.mean(df[col]), inplace=True)

fill_with_mean_values(df)

df.dropna(axis=0, how='any', inplace=True)
y=df['SalePrice']
df.drop(columns=['SalePrice'], axis=1, inplace=True)
X = df
print(X.shape, y.shape)
```

```python
X.sample(5)
```

```python
# Breaking the dataframe into numerical, categorical and target
num_df = X.select_dtypes(include=['float64', 'int64'])
cat_df = X.select_dtypes(include=['object'])
```

```python
# outliers treatment for numerical columns
def outlier_cap(X):
    X=X.clip(lower=X.quantile(0.01))
    X=X.clip(upper=X.quantile(0.99))
    return X

num_df = num_df.apply(lambda x: outlier_cap(x))
```

```python
# Converting Categorical data into numerical data using LabelEncoder
from sklearn.preprocessing import LabelEncoder
cat_cols = cat_df.columns.to_list()
def label_encoding(df):
    for col in df.columns:
        le = LabelEncoder()
        new_col = col+"_lb_enc"
        df[new_col] = le.fit_transform(df[col])
        le = None

label_encoding(cat_df)
cat_df.drop(columns=cat_cols, axis=1, inplace=True)
cat_df.sample(5)       
```

```python
X = pd.concat([num_df, cat_df], axis=1, join='inner')
X.sample(5)
```

```python
X.shape
```

```python
# Remove the features with zero variance 
from sklearn.feature_selection import VarianceThreshold

var_sel = VarianceThreshold(threshold=.8 * (1 - .8))
var_sel.fit_transform(X)
cols=var_sel.get_feature_names_out()
X= X[cols]
X.shape
```

```python
# Train Test split 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=50)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
# Scaling using min-max scalar
from sklearn.preprocessing import MinMaxScaler

mm_sc = MinMaxScaler()
X_train_sc = mm_sc.fit_transform(X_train)
X_test_sc = mm_sc.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error

gbr = GradientBoostingRegressor()
gbr.fit(X_train_sc, y_train)
y_pred = gbr.predict(X_test_sc)

print("Train R2 score: ", gbr.score(X_train_sc , y_train))
print("Test R2 score: ", r2_score(y_test, y_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train_sc, y_train)

y_pred = lr.predict(X_test_sc)

print("Train R2 score: ", lr.score(X_train_sc , y_train))
print("R2 score: ", r2_score(y_test, y_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
```
