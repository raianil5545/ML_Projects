```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('weatherAUS.csv')
df.sample(5)
```

```python
df.info()
```

```python
df.Date = pd.to_datetime(df.Date, format="%Y/%m/%d", errors='raise')
```

```python
df.info()
```

```python
df["Year"]=pd.DatetimeIndex(df.Date).year
df["Month"]=pd.DatetimeIndex(df.Date).month
df["Day"]=pd.DatetimeIndex(df.Date).day
```

```python
df.drop('Date', axis='columns', inplace= True)
df.sample(5)
```

```python
df.info()
```

```python
np.mean(df.MinTemp)
```

```python
def fill_missing_num_values(df):
    for col in df.columns:
        if df[col].dtypes in ["float64", "int64"]:
            df[col].fillna(np.mean(df[col]), inplace=True)

fill_missing_num_values(df)
df.sample(5)
```

```python
df.dropna(axis=0, inplace=True)
df.info()
```

```python
cat_columns = df.columns[df.dtypes=='object'].tolist()
cat_columns.remove("RainTomorrow")
num_columns = df.columns[df.dtypes=='float64'].tolist()
```

```python
cat_columns
```

```python
# checking the values and counts in the object columns
def unique_values_in_object(df, columns):
    for col in columns:
        print(col, df[col].unique())
        print("-x-x-x--x-x-x--x--x-x--x")

unique_values_in_object(df, cat_columns)
```

```python
def removing_categorical_column_as_per_variance(df, columns):
    dropped_columns = []
    for col in columns:
        if len(df[col].unique())>1:
            unique_obj_per = np.array(df[col].value_counts(normalize = True)* 100)
            var = np.var(unique_obj_per)
            if var<1:
                dropped_columns.append(col)
        else:
            dropped_columns.append(col)
    df.drop(dropped_columns, axis='columns', inplace=True)
#removing_categorical_column_as_per_variance(df, cat_columns)
df.head(5)
```

```python
df['RainTomorrow'].value_counts()["Yes"]/len(df['RainTomorrow'])*100
```

```python
cat_columns = df.columns[df.dtypes=='object'].tolist()
cat_columns
```

```python
num_df = df[num_columns]
cat_df = df[cat_columns]
cat_df = pd.get_dummies(cat_df, drop_first=True)
cat_df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
cat_df.sample(5)
y= cat_df["RainTomorrow_Yes"]
cat_df = cat_df.drop("RainTomorrow_Yes", axis='columns')
X = pd.concat([num_df, cat_df], axis=1)
```

```python
X.sample(5)
```

```python
X.shape, y.shape
```

```python
sc=MinMaxScaler()
X_scaled = sc.fit_transform(X)
X_scaled[:5]
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X_scaled, y)
y_sm.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.25, random_state=25, stratify=y_sm)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
# Hyper Parameters Tuning
penalties = ['l1', 'l2']
C = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
solvers = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
l1_ratios = [0.25, 0.5, 0.75]
lgr_best_score = 0
lgr_best_solver = str
lgr_best_c = 0
lgr_best_penalty = str
hyper_params = []
f1_Scores = []
kFolds = 5
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score, confusion_matrix

for s in solvers:
    for p in penalties:
        for c in C:
            if s == 'newton-cg' or 'lbfgs' or 'sag':
                if p =='l2':
                    lgr = LogisticRegression(penalty = p , C = c,  solver = s, max_iter = 100)
                    y_pred = cross_val_predict(lgr, X_train, y_train, cv = kFolds)
                    F1_score = f1_score(y_train, y_pred)
                    hyper_params.append([s,p,c]) 
                    f1_Scores.append(F1_score)
                    if F1_score>lgr_best_score:
                        lgr_best_score = F1_score
                        lgr_best_solver = s
                        lgr_best_c =c
                        lgr_best_penalty = p 
            if s == "liblinear":
                if p in ['l1', 'l2']:
                    lgr = LogisticRegression(penalty = p , C = c,  solver = s, max_iter = 100)
                    y_pred = cross_val_predict(lgr, X_train, y_train, cv = kFolds)
                    F1_score = f1_score(y_train, y_pred)
                    hyper_params.append([s,p,c]) 
                    f1_Scores.append(F1_score)
                    if F1_score>lgr_best_score:
                        lgr_best_score = F1_score
                        lgr_best_solver = s
                        lgr_best_c =c
                        lgr_best_penalty = p  
            if s == "saga":
                if p == "elasticnet":
                    for l1 in l1_ratios:
                        lgr = LogisticRegression(penalty = p , C = c,  solver = s, l1_ratio=l1, max_iter = 100)
                        lgr = LogisticRegression(penalty = p , C = c,  solver = s, max_iter = 100)
                        y_pred = cross_val_predict(lgr, X_train, y_train, cv = kFolds)
                        F1_score = f1_score(y_train, y_pred)
                        hyper_params.append([s,p,c]) 
                        f1_Scores.append(F1_score)
                        if F1_score>lgr_best_score:
                            lgr_best_score = F1_score
                            lgr_best_solver = s
                            lgr_best_c =c
                            lgr_best_penalty = p
                else:
                    lgr = LogisticRegression(penalty = p , C = c,  solver = s, max_iter = 100)
                    y_pred = cross_val_predict(lgr, X_train, y_train, cv = kFolds)
                    F1_score = f1_score(y_train, y_pred)
                    hyper_params.append([s,p,c]) 
                    f1_Scores.append(F1_score)
                    if F1_score>lgr_best_score:
                        lgr_best_score = F1_score
                        lgr_best_solver = s
                        lgr_best_c =c
                        lgr_best_penalty = p 
```

```python
hyper_params
```

```python
index = f1_Scores.index(max(f1_Scores))
hyper_params[index]
```

```python
lgr_best_penalty, lgr_best_c, lgr_best_solver
```

```python
best_lgr = LogisticRegression(penalty = lgr_best_penalty, C = lgr_best_c, solver = lgr_best_solver, max_iter=5000)
best_lgr.fit(X_train, y_train)
```

```python
y_train_predictions = best_lgr.predict(X_train)
```

```python
score = f1_score(y_train, y_train_predictions)
score
```

```python
y_pred = best_lgr.predict(X_test)
y_pred[:5]
```

```python
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test, y_pred, target_names=['0','1']))
print(accuracy_score(y_train, y_train_predictions))
print(accuracy_score(y_test, y_pred))
```
